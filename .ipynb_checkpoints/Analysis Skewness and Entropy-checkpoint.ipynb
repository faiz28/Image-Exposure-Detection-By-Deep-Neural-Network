**------------Normal green image entropy and skewness-----------------**

**Entropy**

```python
# Normal image
Green_Normal_entropy = df[" 'Normal'.1"][2:]
Green_Normal_entropy =  Green_Normal_entropy.astype(float)
ax =  np.arange(0,Green_Normal_entropy.count(),1)

# Overexposed iamge
Overexposed_entropy = df[" 'OverExposed'.1"][2:]
Overexposed_entropy =  Overexposed_entropy.astype(float)

#Underexposed image
Underexposed_entropy = df[" 'UnderExposed'.1"][2:]
Underexposed_entropy =  Underexposed_entropy.astype(float)
```

```python
plt.plot(ax,Green_Normal_entropy)
plt.show()
plt.plot(ax,Overexposed_entropy)
plt.show()
plt.plot(ax,Underexposed_entropy)
plt.show()
```

```python
plt.hist(Green_Normal_entropy,np.arange(0,20,0.5))
plt.show()
plt.hist(Overexposed_entropy,np.arange(0,20,0.5))
plt.show()
plt.hist(Underexposed_entropy,np.arange(0,20,0.5))
plt.show()
```

**Skewness**

```python
# Normal image
Green_Normal_skewness = df[" 'Normal'.2"][2:]
Green_Normal_skewness =  Green_Normal_skewness.astype(float)
ax =  np.arange(0,Green_Normal_entropy.count(),1)

# Overexposed iamge
Overexposed_skewness = df[" 'OverExposed'.2"][2:]
Overexposed_skewness =  Overexposed_skewness.astype(float)

#Underexposed image
Underexposed_skewness = df[" 'UnderExposed'.2"][2:]
Underexposed_skewness =  Underexposed_skewness.astype(float)
```

```python
plt.plot(ax,Green_Normal_skewness)
plt.show()
plt.plot(ax,Overexposed_skewness)
plt.show()
plt.plot(ax,Underexposed_skewness)
plt.show()
```

```python
plt.hist(Green_Normal_skewness,np.arange(2.5,6,0.1))
plt.show()
plt.hist(Overexposed_skewness,np.arange(2.5,6,0.1))
plt.show()
plt.hist(Underexposed_skewness,np.arange(2.5,6,0.1))
plt.show()
```
